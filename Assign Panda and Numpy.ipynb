#Assign Libraries: NumPy & Pandas


```python

import pandas as pd
import numpy as np
df = pd.read_csv('retail_dataset .csv')
```

#Q1. Load the given retail sales dataset into a Pandas DataFrame and display the first 5 rows along with the dataset information to understand its structure.

```python
df.head()
```

#Q2. Inspect the dataset and identify missing values in each column. Display the total number of missing values present in the dataset.

```python
missing_values = df.isnull().sum()
total_missing = df.isnull().sum().sum()

print("Missing values per column: ", missing_values)
print("Total missing values in the dataset: " + str(total_missing))
```

#Q3. Replace missing values in the Quantity and Price columns with their respective mean values using Pandas.

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())
```

#Q4. Remove all rows where the Product Category or Region column contains missing values.

```python
df.dropna(subset=['Product Category', 'Region'], inplace=True)
```

#Q5. Create a new column called Revenue using NumPy, where: Revenue = Quantity × Price Display the updated dataset with the new column.

```python
df['Revenue'] = df['Quantity'] * df['Price']
print("\n Updated Dataset with Revenue")
print(df.head())
```

#Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.

```python
Total_Revenue = df['Revenue'].sum()
print("\n Total Revenue: " + f"${Total_Revenue:,.2f}")
```

#Q7. Group the dataset by Product Category and calculate the total revenue generated by each category.

```python
Category_Revenue = df.groupby('Product Category')['Revenue'].sum()
print("\n Revenue by Product Category:")
print(Category_Revenue)
```

#Q8. Based on the revenue calculated for each product category, identify the top 3 categories with the highest revenue and the bottom 3 categories with the lowest revenue.


```python
Top3_Categories = Category_Revenue.nlargest(3)
Bottom3_Categories = Category_Revenue.nsmallest(3)
print("\n Top 3 Product Categories by Revenue:")
print(Top3_Categories)
print("\n Bottom 3 Product Categories by Revenue:") 
print(Bottom3_Categories)
```

#Q9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the region with the highest revenue and the region with the lowest revenue.


```python
region_revenue = df.groupby('Region')['Revenue'].sum()
highest_revenue_region = region_revenue.idxmax()
lowest_revenue_region = region_revenue.idxmin()
print("\n Region with Highest Revenue: ", highest_revenue_region)
print("Region with Lowest Revenue: ", lowest_revenue_region)
print("\n Revenue by Region:")
print(region_revenue)
```

#Q10. Extract the month from the Date column using Pandas and group the dataset by month to calculate the total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column using NumPy.


```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['Month'] = df['Date'].dt.month


monthly_revenue = df.groupby('Month')['Revenue'].sum()

rev_mean = np.mean(monthly_revenue)
rev_median = np.median(monthly_revenue)
rev_std = np.std(monthly_revenue)
print("\n Monthly Revenue:")
print(monthly_revenue)
print("\n Mean:", rev_mean)
print("Median:", rev_median)
print("Standard Deviation:", rev_std)
```
